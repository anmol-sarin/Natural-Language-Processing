```python
# making a function
def get_match(pattern , text):
    matches = re.findall(pattern , text)
    if matches:
        return matches[0]
    
def personal_info(text):
    name = get_match('Born[^.](.*)', text)
    birth = get_match('Born[^.].*\n(.*)\(',text)
    place = get_match('Born[^.].*\n.*\n(.*)',text)
    return{
        'name' : name,
        'date of birth' : birth,
        'place':place
    }
    
    
```
