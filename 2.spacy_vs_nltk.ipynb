```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')

doc = nlp("Dr. strange loves pav bhaji of mumbai. Hulk loves chaat of Delhi")
# Sentance Tokenization in spacy
# for sentance in doc.sents:
    # print (sentance)
# Work Tokenization in spacy
for sentance in doc.sents:
    for words in sentance:
        print(words)
```
